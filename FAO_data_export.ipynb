```python
from data_config import fao_fra_2025_raw_data, FAO_name_fix

fao_fra_2025_data = fao_fra_2025_raw_data.copy()
fao_fra_2025_data = fao_fra_2025_data.replace(FAO_name_fix)
print(fao_fra_2025_data.head())
year_columns = ['1990_area_Mha', '2000_area_Mha', '2010_area_Mha', '2015_area_Mha', '2020_area_Mha', '2025_area_Mha']
fao_fra_2025_data[year_columns] =  fao_fra_2025_data[year_columns] /1000
print(fao_fra_2025_data.head())

DATA_folder = '/content/Biochar_forest_estimation/data/'

fao_fra_2025_data.to_csv(DATA_folder + 'fao_fra_2025_data.csv', index = False)

print(os.listdir('/content/Biochar_forest_estimation/data/'))
```
